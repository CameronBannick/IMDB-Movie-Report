# Connect to and Update mySQL DB

## Connect to mySQL DB

```python
#connecting to mySQL
connection = "mysql+pymysql://root:root@localhost/movies"
engine = create_engine(connection)
engine
```

```python
#creating database
if database_exists(connection)==False:
    create_database(connection)
else: print('Database already exists')
```

## Load 2011 DataFrame

```python
#inspecting dataframe
tmdb_2011.info()
```

```python
#saving TMDB data to my sq
tmdb_2011.set_index('imdb_id').to_sql('tmdb_data', engine,  if_exists='append', index=True)
```

```python
#test query to see total entries to make sure its update
q ="""SELECT *
      FROM tmdb_data;"""

pd.read_sql(q, engine)
```
